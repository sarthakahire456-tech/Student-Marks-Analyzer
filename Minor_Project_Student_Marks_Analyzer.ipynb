# Minor Project: Student Marks Analyzer using Python and NumPy

## Problem Statement
The objective of this project is to analyze student marks using Python and NumPy. The program takes marks as input, stores them in an array, and calculates total, average, highest, and lowest marks. Based on the average, it assigns a grade using conditional statements.

## Logic
1. Take number of students as input
2. Use a loop to take marks
3. Store marks in a list
4. Convert list to NumPy array
5. Calculate total, average, max, and min
6. Use if-elif to assign grade
7. Display results

```
import numpy as np

# Function to assign grade
def assign_grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "Fail"

# Main program
n = int(input("Enter number of students: "))

marks_list = []

for i in range(n):
    mark = int(input(f"Enter marks for student {i+1}: "))
    marks_list.append(mark)

# Convert to NumPy array
marks_array = np.array(marks_list)

# Calculations
total = np.sum(marks_array)
average = np.mean(marks_array)
highest = np.max(marks_array)
lowest = np.min(marks_array)

grade = assign_grade(average)

# Output
print("\n--- Results ---")
print("Marks:", marks_array)
print("Total:", total)
print("Average:", average)
print("Highest:", highest)
print("Lowest:", lowest)
print("Grade:", grade)
```
